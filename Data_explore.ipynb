## Check covalence, correlation 

```python
print(df_parkinson.cov())
```

```python
print(df_parkinson.corr())
```

```python
plt.rcParams["figure.figsize"] = (25, 20)
plt.title('Correlation matrix of  features')
sns.heatmap(df_parkinson.corr(), cmap="coolwarm", annot=True)
```

```python
plt.rcParams["figure.figsize"] = (25, 20)
plt.title('Covalence matrix of  features')
sns.heatmap(df_parkinson.cov(), cmap="Blues", annot=True)
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df_parkinson, figsize=(25,25))
```

```python
lst=['name','HNR','NHR','RPDE','DFA','D2']
df_parkinson_n= df_parkinson.drop(lst, 1, errors='ignore')
df_parkinson_n.head()
df_parkinson_n.info()
```

```python
fig, ax = plt.subplots(6, 3, figsize=(35,25))
ndarray_disease_features = df_parkinson_n.columns.values.reshape(6, 3)
print(ndarray_disease_features )
for i in range(6):
    for j in range(3):
        str_feature = ndarray_disease_features [i, j]
        df_parkinson_n[str_feature].plot(kind="hist", bins=30, ax=ax[i, j])
        ax[i, j].set(title=str_feature)
```
